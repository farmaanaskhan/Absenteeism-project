```python
import pandas as pd
```

```python
df=pd.read_csv('Absenteeism_data.csv')
```

```python
data=df.copy()
```

```python
data
```

```python
data.drop(['Age'], axis=1)
data
```

```python
columns=pd.get_dummies(df['Age'])
```

```python
columns
```

```python
col1=columns.loc[:,27:39].max(axis=1)
col2=columns.loc[:,40:45].max(axis=1)
col3=columns.loc[:,46:50].max(axis=1)
col4=columns.loc[:,51:58].max(axis=1)
```

```python
data=pd.concat([data,col1,col2,col3,col4],axis=1)
data
```

```python
data.columns.values
```

```python
column_names=['ID', 'Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'loop_1','loop_2','loop_3','loop_4']
```

```python
data.columns=column_names
```

```python
column_names_reorder=[ 'loop_1','loop_2','loop_3','loop_4','ID', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours','Reason for Absence']
```

```python
data['Date']=pd.to_datetime(data['Date'],format='%d/%m/%Y')
```

```python
list_months=[]
```

```python
for i in range(data.shape[0]):
    list_months.append(data['Date'][i].month)
```

```python
data['Month_value']=list_months
```

```python
data['Date'][0].weekday()
```

```python
def week_day_from_date(date_value):
    return date_value.weekday()
```

```python
data['weekday']=data['Date'].apply(week_day_from_date)
```
